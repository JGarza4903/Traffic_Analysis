```python
drop_df = clark_df.drop(['OBJECTID','County','Property_Damage_Only','Agency','CrashDateTimeUTC','GlobalID','Fatalities','V1_Driver_Distracted','V1_Most_Harmful_Event','V1_All_Events','V2_Driver_Distracted','V2_Most_Harmful_Event','V2_All_Events','Nonmotorist_Factors','Distance','Injured','V1_Lane_Num','V2_Lane_Num','V2_Vehicle_Factors','First_Harmful_Event','Accident_Rec_Num'], axis=1)
```

```python
drop_df.head()
```

```python
drop_df.shape
```

```python
drop_df.isnull().sum()
```

```python
drop_df['Pedalcyclist'].fillna('Unknown', inplace=True)
drop_df['Pedestrian'].fillna('Unknown', inplace=True)
drop_df['Motorcyclist'].fillna('Unknown', inplace=True)
drop_df['AnimalType'].fillna('Unknown', inplace=True)
drop_df['Secondary_Street'].fillna('Unknown', inplace=True)
drop_df['Weather'].fillna('Unknown', inplace=True)
drop_df['Injury_Type'].fillna('Unknown', inplace=True)
drop_df['Crash_Type'].fillna('Unknown', inplace=True)
drop_df['V1_Type'].fillna('Unknown', inplace=True)
drop_df['V1_Dir'].fillna('Unknown', inplace=True)
drop_df['V1_Action'].fillna('Unknown', inplace=True)
drop_df['V1_Driver_Factors'].fillna('Unknown', inplace=True)
drop_df['V1_Vehicle_Factors'].fillna('Unknown', inplace=True)
drop_df['V2_Type'].fillna('Unknown', inplace=True)
drop_df['V2_Dir'].fillna('Unknown', inplace=True)
drop_df['V2_Action'].fillna('Unknown', inplace=True)
drop_df['V2_Driver_Factors'].fillna('Unknown', inplace=True)
drop_df['Factors_Roadway'].fillna('Unknown', inplace=True)
drop_df['Lighting'].fillna('Unknown', inplace=True)
drop_df['HWY_Factors'].fillna('Unknown', inplace=True)
```

```python
drop_df.isnull().sum()
```

```python
drop_df['V1_Driver_Age'] = drop_df['V1_Driver_Age'].fillna(drop_df['V1_Driver_Age'].mean())
```

```python
drop_df.isnull().sum()
```

```python
drop_df['V2_Driver_Age'] = drop_df['V2_Driver_Age'].fillna(drop_df['V2_Driver_Age'].mean())
```

```python
drop_df.isnull().sum()
```

```python
drop_df.duplicated().sum()
```

```python
dupe_df = drop_df.drop_duplicates()
dupe_df.head(5)
```

```python
dupe_df.duplicated().sum()
```

```python
dupe_df.dtypes
```

```python
dupe_df['Crash_Date'] = pd.to_datetime(dupe_df['Crash_Date'], format = '%Y-%m-%d %H:%M')
dupe_df.dtypes
```

```python
dupe_df['Crash_Time'] = pd.to_datetime(dupe_df['Crash_Date'], format = '%Y-%m-%d %H:%M')
dupe_df.dtypes
```

```python
dupe_df['V1_Driver_Age']
```

```python
dupe_df['V1_Driver_Age'] = dupe_df['V1_Driver_Age'].astype(int)
dupe_df.dtypes
```

```python
dupe_df['V2_Driver_Age'] = dupe_df['V2_Driver_Age'].astype(int)
dupe_df.dtypes
```
